```python
import re  
  
# Natural Language Tool Kit 
import nltk  
  
# to remove stopword 
from nltk.corpus import stopwords 
  
# for Stemming propose  
from nltk.stem.porter import PorterStemmer 
  
# Initialize empty array to append clean text 
corpus = []  
  
# 1000 (reviews) rows to clean 
for i in range(0, 5572):  
      
    # column : "Review", row ith 
    review = re.sub('[^a-zA-Z]', ' ', original_df['v2'][i])  
      
    # convert all cases to lower cases 
    review = review.lower()  
      
    # split to array(default delimiter is " ") 
    review = review.split()  
      
    # creating PorterStemmer object to 
    # take main stem of each word 
    ps = PorterStemmer()  
      
    # loop for stemming each word 
    # in string array at ith row     
    review = [ps.stem(word) for word in review 
                if not word in set(stopwords.words('english'))]  
                  
    # rejoin all string array elements 
    # to create back into a string 
    review = ' '.join(review)
     
         
    # append each string to create 
    # array of clean text  
    corpus.append(review) 
```

```python
len(corpus)
```

```python
original_df['new_Phrase']=corpus
```

```python
# list of text documents
from sklearn.feature_extraction.text import TfidfVectorizer
bag=original_df['new_Phrase']
# create the transform
vectorizer1 = TfidfVectorizer(norm=None,analyzer='word')
# tokenize and build vocab
vectorizer1.fit(bag)
# summarize
print(vectorizer1.vocabulary_)
print(vectorizer1.idf_)
# encode document
vector1 = vectorizer1.transform(bag)
# summarize encoded vector
print(vector1.shape)
print(vector1.toarray())
bow_tfidf = vector1.toarray()
```

```python
bag_df_tfidf = pd.DataFrame(bow_tfidf)
bag_df_tfidf
```

```python
from sklearn.model_selection import train_test_split
X_train_desc_tfidf, X_test_desc_tfidf, y_train_desc_tfidf, y_test_desc_tfidf = train_test_split(bag_df_tfidf, dummy_ham, test_size = 0.20,random_state=50) 
```

```python
from sklearn.linear_model import LogisticRegression 
logreg_desc_tiidf = LogisticRegression()
logreg_desc_tiidf.fit(X_train_desc_tfidf,y_train_desc_tfidf)
logreg_desc_tiidf.score(X_test_desc_tfidf, y_test_desc_tfidf)
```

```python
from sklearn.tree import DecisionTreeClassifier
decTree_desc_tiidf = DecisionTreeClassifier()
decTree_desc_tiidf.fit(X_train_desc_tfidf,y_train_desc_tfidf)
decTree_desc_tiidf.score(X_test_desc_tfidf, y_test_desc_tfidf)
```

```python
from sklearn.metrics import confusion_matrix
decTree_desc_tiidf_ypred =decTree_desc_tiidf.predict(X_test_desc_tfidf)
cm=confusion_matrix(y_test_desc_tfidf,decTree_desc_tiidf_ypred)
cm
```

```python
accuracy = (cm[0,0]+cm[1,1])/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1])
recall = cm[0,0]/(cm[0,0]+cm[0,1])
precision = cm[0,0]/(cm[0,0]+cm[1,0])
print(recall)
```

```python
F1 = 2 * (precision * recall) / (precision + recall)
F1
```
