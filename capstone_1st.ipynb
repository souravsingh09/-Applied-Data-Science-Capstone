```python
import requests
wiki='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
link=requests.get(wiki).text
```

```python
#link
```

```python
from bs4 import BeautifulSoup
soup= BeautifulSoup(link,'lxml')
#soup
```

```python
#soup.title.string
```

```python
all_a = soup.find_all("a")
for link in all_a:
    print(link.get("href"))
```

```python
find_tab=soup.find_all('table')
```

```python
find_tables= soup.find('table',class_='wikitable sortable')
print(find_tables)
```

```python
import pandas as pd
table_tr = find_tables.find_all("tr")
columns=[v.text.replace('\n','') for v in table_tr[0].find_all('th')]
columns
```

```python
df=pd.DataFrame(columns=columns)
for i in range(1,len(table_tr)):
    tds=table_tr[i].find_all('td')
    
    if len(tds) == 3:
        values=[tds[0].text,tds[1].text, tds[2].text.replace('\n','')]
    else:
        values =[tds.text for td in tds]
    
    df = df.append(pd.Series(values,index=columns),ignore_index = True)
```
